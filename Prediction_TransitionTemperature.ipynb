```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ChainLength = np.array([12,14,16,18])
TransitionTemperature = np.array([-2,24,41,55])
Temp_array = np.arange(-2,55,1)
Temp_array
```

```python
c = np.polyfit(ChainLength,TransitionTemperature,1)
c
```

```python
ChainPred = (Temp_array - c[1])/c[0]
```

```python
fig,ax = plt.subplots(figsize=(10,5))

ax.plot(ChainLength, TransitionTemperature,'.')
ax.plot(ChainPred, Temp_array,'--')
```
