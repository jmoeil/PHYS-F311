```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ChainLength = np.array([12,14,16,18])
TransitionTemperature = np.array([-2,24,41,55])
Temp_array = np.arange(-2,55,1)
Temp_array
```

```python
c = np.polyfit(ChainLength,TransitionTemperature,1)
c
```

```python
ChainPred = (Temp_array - c[1])/c[0]
```

```python
fig,ax = plt.subplots(figsize=(10,5))

ax.plot(ChainLength, TransitionTemperature,'.',label='$T_m$ from litterature')
ax.plot(ChainPred, Temp_array,'--',label='Linear regression on litterature data')
ax.plot(14,23.512112024083997,'^',label='Densitometer DMPC $T_m$')
ax.plot(16,41.39048545117069,'^',label='Densimeter DPPC $T_m$')

ax.set_xlabel('Chain length [# C atoms]')
ax.set_ylabel('Temperature [°C]')

ax.legend()
```
