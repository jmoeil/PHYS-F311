```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ADC_50_ns = np.round(np.mean(np.genfromtxt('Calibration/ADC_50_ns',comments='#')))
ADC_200_ns = np.round(np.mean(np.genfromtxt('Calibration/ADC_200_ns',comments='#')))
ADC_1_us = np.round(np.mean(np.genfromtxt('Calibration/ADC_1_us',comments='#')))
ADC_10_us = np.round(np.mean(np.genfromtxt('Calibration/ADC_10_us',comments='#')))
ADC_5_us = np.round(np.mean(np.genfromtxt('Calibration/ADC_5_us',comments='#')))
ADC_2dot5_us = np.round(np.mean(np.genfromtxt('Calibration/ADC_2dot5_us',comments='#')))
ADC_7dot5_us = np.round(np.mean(np.genfromtxt('Calibration/ADC_7dot5_us',comments='#')))

ADC = np.array([ADC_50_ns,ADC_200_ns,ADC_1_us,ADC_10_us,ADC_5_us,ADC_2dot5_us,ADC_7dot5_us])
```

```python
I = 1.15
dt = np.array([50e-9,200e-9,1e-6,10e-6,5e-6,2.5e-6,7.5e-6])
Q = dt*I
c=np.polyfit(ADC,Q,1)
ADC,Q
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,5))

ax.plot(ADC,Q,'rp')
ax.plot(ADC,c[0]*ADC+c[1],'--')
ax.set_xlabel('ADC values')
ax.set_ylabel('Charge Q [C]')
ax.set_title('Linear relation between the charge Q and ADC values')
```
