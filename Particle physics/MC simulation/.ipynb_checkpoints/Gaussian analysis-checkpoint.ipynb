```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp
```

```python
def gaussian(x,mu,sigma):
    
    f=1/(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/(2*sigma**2))
    
    return f
```

```python
x = np.arange(0,14,0.1)
mu,sigma = 7,1
```

```python
fig, ax = plt.subplots(figsize=(7,7))

ax.plot(x,gaussian(x,mu,sigma))

```

```python
y=np.linspace(-1,1,1000) #l'inverse de scipy_cumul_gauss a un domaine compris entre -1 et 1 chez scipy

inv_cumul_gauss = scp.special.erfinv(y)

fig, ax = plt.subplots(figsize=(7,7))

ax.plot(y,inv_cumul_gauss)
```

```python
mu=5
sigma=4
```

```python
rnd=np.random.uniform(-1,1,10000)

inv_cumul_gauss = scp.special.erfinv(rnd)

fig, ax = plt.subplots(figsize=(7,7))

ax.hist(inv_cumul_gauss*sigma*np.sqrt(2)+mu, bins=40)

```
